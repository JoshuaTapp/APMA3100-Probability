```python
import math

def exponential(lamda, x):
    if(lamda > 0 and x > 0): # Domain restriction
        return lamda * math.exp(-1*lamda*x) # This is the pdf of Exponential() => f(x) = e^(-x)
    else:
        return -1 # Error catching
exponential(1/12, 1)
```

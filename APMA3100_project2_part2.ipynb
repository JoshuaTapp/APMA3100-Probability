```python
import math
import matplotlib.pyplot as plt
# From https://en.wikipedia.org/wiki/Inverse_transform_sampling
# F^-1(r), where r is continous uniform random variable (0,1)
# and F^-1 = -1/lambda * ln(1-r)
def exponential(r):
    return (-12*math.log(1-r)) 

def randomGen():
    seed = 1000
    a = 24693
    c = 1753
    K = 32768
    while True:
        seed = (((a * seed) + c) % K) # update seed
        output = seed / K             # return the random number
        yield output
        
# Monte Carlo Simulation
rand = randomGen()
def Monte_Carlo_sim(n):
    times=[] # list to hold the times for each simulation.
   
    # Run the simulation
    for i in range(n):
        times.append(callSim())

    # Calculate Stats
    mean = 0
    for x in times:
        mean += x
    
    times.sort()
    print("Mean: ", mean/n, "\tFirst Quartile: %3.4f" %times[249], "\tMedian: %3.4f" %times[499], "\tThird Quartile: %3.4f" %times[749])
    
    
    #Calculate Probabilities
    p15 = searchList(times, 15)/1000
    p20 = searchList(times, 20)/1000
    p30 = searchList(times, 30)/1000
    p40 = 1 - searchList(times, 40)/1000
    p65 = 1 - searchList(times, 65)/1000
    p90 = 1 - searchList(times, 90)/1000
    p100 = 1 - searchList(times, 100)/1000
    print("F(15) = %3.4f" %p15, "\nF(20) = %3.4f" %p20, "\nF(30) = %3.4f" %p30) 
    print("1 - F(40) = %3.4f" %p40, "\n1 - F(65) = %3.4f" %p65, "\n1 - F(90) = %3.4f" %p90, "\n1 - F(100) = %3.4f" %p100)
    print()
    
    # Print Graph of CDF of W
    plt.plot([7, 15, 20, 30, 40, 65, 90, 100, 108], [0, p15, p20, p30, (1-p40), (1-p65),(1-p90), (1-p100), 1])
    plt.xlabel("w")
    plt.ylabel("F(w)")
    plt.title("CDF of Random Variable W")
    plt.show()

# The individual simulation of each call    
def callSim():
    call=0
    time=0
    while(call < 4):
        
        if(next(rand) >= 0.5): # Caller Pick ups with 50% prob.
            time+=6 # time to turn on phone and call number
            t = exponential(next(rand))
            if (t > 20):
                time+=21  # treat as missed call
                call+=1    # call increase
                continue  # call failed, so lets continue sim.
            else:
                time+=t   # time += exponential(next(rand)) => value = [0,20]
                time+=1   # add time to hang up
                call+=1   # call increase
                break     # we have a completed call, break out of loop. (end this simulation)
            
        else:
            time+=6
            
            if(next(rand) >= 0.4): # no pick up
                time+=20
                
            else:  # busy
                time+= 5
                
            # End Call - add values
            time+=1
            call+=1
    
    # When call >= 4 or we got a pickup
    #print(time)
    return time

# linear search used to find first occurance of a value greater than x. This will give us the
# index of the value that is approximately = F_W(w)
def searchList(list, x):
    for i in range(len(list)):
        if list[i] > x:
            return i
    
    return -1 #used for error catching.
        
    
    
        
Monte_Carlo_sim(1000)    
            
```

# Analysis

### Shape of the Distribution

With our simulation, we have a mean of 40.7 and a median of 34.5. This suggest that there are a higher amount of completed simulation that are less than the mean. This would make the distribution large on the left side, with a slower decay on the right. We know this is true, because although 75% of simulations were under 63 seconds, while the other 25% happened somewhere between 63 and 108 seconds. 

### Sample Space

The sample space could have been deduced before the simulation, and has been confirmed afterwords. With us having the callee pickup the phone on the first try and time to pick up the phone approaching the limit of zero, we would have our lower at seven seconds. While for our upper bonds, because our exponential distribution has the range of $(0,\infty)$, our upper limit is also infinity. Therefore the Sample Space is $ S = { w \ge 7 } $

### Graph of the CDF of W

The above graph will update based on the number of simulations or a new random number generator.

### Is W a true exponential variable

No, for the following reasons:
1. The domain of W is $(7,\infty)$. However the domain of an exponential variable is $[0,\infty)$.
2. The exponential function is required to be memoryless, however our model has memory as a key function. The memory of how many calls have been made.
3. If we take the quartile function, we arrive at a $\lambda = \frac{1}{52.499}$ from the calculation of our first quartile value. If we use this $\lambda$ value for our median and third quartile, we have the following. 

$$Median = 52.499\cdot ln(2) = 36.389$$ 
$$Third \  Quartile = 52.599 \cdot ln(4) = 72.778$$

From this, we can see, the distribution does not follow the exponential distribution.
